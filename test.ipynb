```python
import torch
def euclidean_dist(x, y):
    # x: N x D
    # y: M x D
    n = x.size(0)
    m = y.size(0)
    d = x.size(1)
    assert d == y.size(1)

    x = x.unsqueeze(1).expand(n, m, d)
    y = y.unsqueeze(0).expand(n, m, d)

    score = -torch.pow(x - y, 2).sum(2)
    return score

x = torch.rand([15,100])
y = torch.rand([5,100])
euclidean_dist(x,y).shape
```
